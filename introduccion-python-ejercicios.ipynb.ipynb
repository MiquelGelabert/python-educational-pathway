# Raiz cubica

```python
def raizcubica(n):
    print('Buscador de raíces cúbicas perfectas')
    raiz = 0
    while raiz**3 < abs(n):
        raiz += 1

    if raiz**3 == abs(n):
        if n<0:
            print(-raiz)
        else:
            print(raiz)
    else:
        print('No tiene raiz cúbica perfecta')
raizcubica(64)
```

# Factorial

```python
def factorial(x):
    if x == 0:
        return 1
    else:
        return x*factorial(x-1)
print(factorial(10))
```

```python
def lista_factorial(n): #factorial a partir del último elemento de la lista
    l=[1, 1]
    for k in range(2,n+1):
        a = k*l[-1]
        l.append(a)
    return l
lista_factorial(20)
```

```python
import matplotlib.pyplot as plt
def cifras_factorial(n): #cuantas cifras tiene cada factorial
    l=[1, 1]
    t= []
    for k in range(2,n+1):
        a = k*l[-1]
        l.append(a)
    for element in l:
        t.append(len(str(element)))
    plt.step(range(n+1),t)
cifras_factorial(100)
```

# Persistencia multiplicativa

La persistencia multiplicativa de un número es la cantidad de pasos requeridos para reducirlo a una cifra multiplicando sus dígitos. Por ejemplo, la persistencia de 39 es 3 porque 3×9 = 27, 2×7 = 14 y 1×4 = 4.

```python
def product(n):
  if len(str(n)) == 1:
    return n
  multiplication = 1
  for i in str(n):
    multiplication = int(i)*multiplication
  return multiplication
```

```python
def multi_per(n):
  iterations = 0
  while len(str(n)) != 1:
    n = product(n)
    iterations += 1
  return iterations
```

```python
def max_iterations_till_N(N):
  max_iterations = 0
  number = 0
  for i in range(N//10, N):
    if multi_per(i) > max_iterations:
      max_iterations = multi_per(i)
      number = i
  return max_iterations, number

max_iterations_till_N(10000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_values = np.unique(np.rint(np.logspace(np.log10(100), np.log10(1_000_000), num=50)).astype(int))
print(f"Generated {len(n_values)} logarithmically spaced N values, from {n_values.min()} to {n_values.max()}")

max_persistence = 0
num_at_max_persistence = 0
running_n_values = []
running_max_pers = []
running_num_at_max_pers = []

for N in n_values:
    # Call max_iterations_till_N for the current N
    current_max_pers, current_num = max_iterations_till_N(N)

    # Update overall max_persistence if a new higher one is found
    if current_max_pers > max_persistence:
        max_persistence = current_max_pers
        num_at_max_persistence = current_num

    # Store the running maximums for the current N
    running_n_values.append(N)
    running_max_pers.append(max_persistence)
    running_num_at_max_pers.append(num_at_max_persistence)

print(f"Finished collecting running maximums for {len(n_values)} N values.")
print(f"Final max persistence found: {max_persistence} by number {num_at_max_persistence}")

fig, ax = plt.subplots(figsize=(12, 7))

# Plot the running maximum persistence
ax.plot(running_n_values, running_max_pers, marker='o', linestyle='-', color='skyblue', label='Max Persistence')

# Set x-axis to logarithmic scale
ax.set_xscale('log')

# Add title and labels
ax.set_title('Running Maximum Multiplicative Persistence vs. N', fontsize=16)
ax.set_xlabel('N (Upper Limit)', fontsize=12)
ax.set_ylabel('Maximum Persistence', fontsize=12)

# Add grid
ax.grid(True, which="both", ls="-", alpha=0.7)

# Annotate points where a new maximum persistence is found
previous_max_pers = -1
for i in range(len(running_n_values)):
    if running_max_pers[i] > previous_max_pers:
        ax.annotate(
            f'Num: {running_num_at_max_pers[i]} (Pers: {running_max_pers[i]})',
            (running_n_values[i], running_max_pers[i]),
            textcoords="offset points",
            xytext=(5,5),
            ha='left',
            fontsize=9,
            color='red'
        )
        previous_max_pers = running_max_pers[i]

plt.legend()
plt.tight_layout()
plt.show()
```

# Gravitación newtoniana
Calcularemos la fuerza total ejercida sobre un objeto con masa $M = 3 kg$ que está influenciado por 100 objetos con masas $m_{1}, m_{2}, . . . , m_{100}$ . El objeto i-ésimo tiene masa $m_i = (\frac{i}{10}+α)$ kg y se encuentra a una distancia $r_i = \sqrt{\frac{i}{6}+β}$ metros.

La fuerza total la calcularemos como la suma de todas las fuerzas y como la única presente es la gravitatoria, utilizarenos estas dos fórmulas:
$$F_g=G\frac{Mm}{r^2}$$
$$F_{total}= \sum_{i=1}^{100} F_{i} = F_{1}+F_{2}+· · ·+F_{100}$$

Donde: $G = 6.674 × 10^{−11} m^3kg^{-2}s^{-1}, α = 2.1 kg, β = 10−8 m$

```python
def grav(N, M, a, b):
    G = 6.674e-11
    sum = 0
    for i in range(1, N+1):
        m = (i/10)+a
        r = (((i/6)**2)+b)**(1/2)
        sum += (m/(r**2))
    print("La fuerza total ejercida sobre M por las 100 masas es de F=", f"{G*M*sum}", "N, aproximadamente 2.85x10^(-8) N.")
grav(100,3,2.1,1e-8)
```

# Convertidor de sistema decimal a binario y a hexadecimal

```python
def convertidor_decimal():
    n0 = -1
    while n0 < 0:
        try:
            n0 = int(input("Introduce un número natural: "))
            if n0 < 0:
                print("Debes introducir un número natural.")
        except ValueError:
            print("Debes introducir un número entero.")
    binario_i = ""
    n=n0
    while n > 0:
        residuo = n % 2
        binario_i = str(residuo) + binario_i
        n //= 2
    binario_i = binario_i
    print("En sistema binario,", n0, "se escribe: ", binario_i)
    binario_i = str(binario_i)
    while len(binario_i) % 4 != 0:
        binario_i = "0" + binario_i
    equivalencias = {"0000": "0","0001": "1","0010": "2","0011": "3","0100": "4","0101": "5","0110": "6","0111": "7","1000": "8","1001": "9","1010": "A","1011": "B","1100": "C","1101": "D","1110": "E","1111": "F"}
    hexadecimal_i = ""
    for i in range(0, len(binario_i), 4):
        grupo = binario_i[i:i+4]
        valor_hexadecimal = equivalencias[grupo]
        hexadecimal_i += valor_hexadecimal
    print("En sistema hexadecimal,", n0, "se escribe: ", hexadecimal_i)

convertidor_decimal()
```

# Oscilador

```python
import matplotlib.pyplot as plt
import numpy as np

def oscilador(A, ω, δ, dt):
    t = np.arange(0,1,dt)
    k = (0.1, 0.4, 0.9, 1.1)
    plt.figure(figsize=(10,6))
    for i in k:
        x = A*np.exp(-i*ω*t)*np.cos(ω*t+δ)
        plt.plot(t, x, label=f"k={i}")
        plt.xlabel("t (s)")
        plt.ylabel("x (m)")
        plt.title("Oscilador")
        plt.legend()

oscilador(0.1, 20*np.pi, np.pi/8, 1e-3)
```

# Helecho de Barnsley

```python
import matplotlib.pyplot as plt
import random

def barnsley(N):

    x, y = 0, 0
    puntos = [(x, y)]

    for _ in range(N):
        r = random.random()

        if r < 0.85:
            next_x = 0.85 * x + 0.04 * y
            next_y = -0.04 * x + 0.85 * y + 1.6
        elif r < 0.92:
            next_x = 0.2 * x - 0.26 * y
            next_y = 0.23 * x + 0.22 * y + 1.6
        elif r < 0.99:
            next_x = -0.15 * x + 0.28 * y
            next_y = 0.26 * x + 0.24 * y + 0.44
        else:
            next_x = 0
            next_y = 0.16 * y

        x, y = next_x, next_y
        puntos.append((x, y))

    x_coords, y_coords = zip(*puntos)

    plt.figure(figsize=(8, 8))
    plt.scatter(x_coords, y_coords, s=1, c='green')
    plt.axis('off')
    plt.show()

barnsley(100000)
```

# Números catalanes

```python
from math import factorial
def catalan(k):
    l=[]
    for n in range(1,k+1):
        c = int(factorial(2*n)/(factorial(n+1)*factorial(n)))
        l.append(c)
    print(l)
catalan(10)
```

# Constante de Kaprekar

```python
def kaprekar():
    k = input("Ingrese un número de cuatro dígitos: ")
    cont = 0
    while k != 6174:
        lista_digitos = list(str(k))
        orden_mayor_menor = sorted(lista_digitos, reverse=True)
        orden_menor_mayor = sorted(lista_digitos, reverse=False)
        ordenado_mayor = int("".join(orden_mayor_menor))
        ordenado_menor = int("".join(orden_menor_mayor))
        k = ordenado_mayor - ordenado_menor
        cont += 1
    print(k, cont)
kaprekar()
```

# Constante Viswanath

```python
import random
import numpy as np

def viswanath(N, M):
    u = np.zeros((M, N))
    for m in range(M):
        u[m, 0] = 0
        u[m, 1] = 1
        for n in range(2, N):
            if random.random() < 0.25:
                p1 = + 1
                p2 = + 1
            elif 0.25 <= random.random() < 0.5:
                p1 = +1
                p2 = -1
            elif 0.5 <= random.random() < 0.75:
                p1 = -1
                p2 = +1
            else:
                p1 = -1
                p2 = -1
            u[m, n] = p1*u[m, n-1] + p2*u[m, n-2] #fibonacci con mas o menos en los dos términos con 50% cada signo
    v_list = []
    k_list = []
    for k in range(1, N):
        columna = [fila[k] for fila in u]
        v_list.append((abs(int(sum(columna)/len(columna))))**(1/k)) #raiz enésima (iteración) de la media de todos los valores
        k_list.append(k)
    return k_list, v_list
```

```python
import numpy as np
import math as math
def series(N):
    sum=0
    for n in range(0,N+1):
        sum += ((-1)**(n))/(2*n+1)
    print("La serie de Leibniz resulta:",f"{sum:0.5f}", "y multiplicada por 4:",4*sum, "aproximadamente pi")
    sum=0
    for n in range(1,N+1):
        sum += 1/((n*(n+1))/2)
    print("La serie de los inversos de los números triangulares resulta:",f"{sum:0.5f}", "aproximadamente 2")
    sum=0
    for n in range(0,N+1):
        sum += 1/2**n
    print("La serie de los inversos de las potencias de dos resulta:",f"{sum:0.5f}", "aproximadamente 2")
    sum=0
    for n in range(0,N+1):
        sum += (-1)**n/2**n
    print("La serie de los inversos de las potencias de dos alternada resulta:",f"{sum:0.5f}", "aproximadamente 2/3")
    sum=0
    for n in range(0,N+1):
        sum += 1/(factorial(n))
    print("La serie de los inversos de los factoriales resulta:",f"{sum:0.5f}", "aproximadamente e")
    sum=0
    for n in range(1,N+1):
        sum += 1/n**2
    print("La serie de los inversos de los cuadrados resulta:",f"{sum:0.5f}", "multiplicada por 6 y haciendo la raiz cuadrada:",(6*sum)**0.5, "aproximadamente pi")
    sum=2
    for n in range(1,N+1):
        sum += 1/((1/math.sqrt(5))*(((1+math.sqrt(5))/math.sqrt(2))**n-((1-math.sqrt(5))/math.sqrt(2))**n))
    print("La serie de los inversos de los números de fibonacci resulta:",f"{sum:0.5f}", "aproximadamente la constante de los inversos de fibonacci")
    sum=0
    for n in range(1,N+1):
        sum += (-1)**(n+1)/n
    print("La serie de los inversos de los naturales alternada resulta:",f"{sum:0.5f}", "aproximadamente ln2")
    sum=0
    for n in range(0,N+1):
        sum += (factorial(2*n)/(factorial(n)*factorial(2*n-n)))/((2*n+1)*16**n)
    print("La serie del algoritmo de Newton resulta:",f"{sum:0.5f}", "multiplicada por 3:",(3*sum), "aproximadamente pi")

series(10)
```

# Parametrización de curvas

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el parámetro r para controlar la apertura de la espiral
r = 0.3

# Crear valores de parámetro (t) en un rango
t = np.linspace(0, 4 * np.pi, 100)  # Generamos 100 puntos en el rango [0, 4π]

# Calcular las coordenadas (x, y) usando la parametrización
x = r * t * np.cos(t)
y = r * t * np.sin(t)

# Dibujar la espiral de Arquímedes
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.title('Espirál de Arquímedes')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.axis('equal')  # Hace que los ejes tengan la misma escala
plt.show()
```

# Derivada e integral de una función

```python
from sympy import *
import sympy as sym
x = Symbol('x')
def funcion(f):
    deri= diff(f,x)
    inte= integrate(f, x)
    print("Para la función f(x)=",f,"su derivada es f'(x)=",deri)
    print("Para la función f(x)=",f,"su integral indefinida es F(x)=",inte)
    p1 = plot(f, legend = True, line_color="red", xlabel="x", ylabel="f(x)",show = False)
    p1[0].label = f
    p2 = plot(deri, legend = True, line_color="green", show = False)
    p2[0].label = deri
    p3 = plot(inte, legend = True, line_color="blue", show = False)
    p3[0].label = inte
    p1.append(p2[0])
    p1.append(p3[0])
    plt.grid(True)
    p1.show()
funcion((x*(1-x)/10))
```

# Números romanos

```python
# https://chat.openai.com/chat
def decimal_to_roman(decimal):
    # Creamos un diccionario con los valores de cada símbolo romano
    roman_dict = {
        1: 'I', 4: 'IV', 5: 'V', 9: 'IX', 10: 'X',
        40: 'XL', 50: 'L', 90: 'XC', 100: 'C',
        400: 'CD', 500: 'D', 900: 'CM', 1000: 'M'
    }

    # Creamos una lista vacía para almacenar los símbolos romanos
    roman_numeral = []

    # Mientras el número decimal sea mayor que cero,
    # seguimos buscando el símbolo romano más grande que podamos usar
    while decimal > 0:
        # Buscamos el símbolo romano más grande que sea menor o igual al número decimal
        for key in sorted(roman_dict.keys(), reverse=True):
            if key <= decimal:
                # Añadimos el símbolo romano a la lista
                roman_numeral.append(roman_dict[key])
                # Restamos el valor del símbolo romano al número decimal
                decimal -= key
                # Salimos del bucle
                break

    # Unimos todos los símbolos romanos en una cadena y la devolvemos
    return ''.join(roman_numeral)

def roman_to_decimal(roman):
    # Creamos un diccionario con los valores de cada símbolo romano
    roman_dict = {
        'I': 1, 'IV': 4, 'V': 5, 'IX': 9, 'X': 10,
        'XL': 40, 'L': 50, 'XC': 90, 'C': 100,
        'CD': 400, 'D': 500, 'CM': 900, 'M': 1000
    }

    # Creamos una variable para almacenar el resultado
    result = 0

    # Iteramos a través de los símbolos romanos en orden inverso
    for i in range(len(roman)-1, -1, -1):
        # Si el símbolo actual es mayor que el siguiente, lo añadimos al resultado
        if i == len(roman)-1 or roman_dict[roman[i]] >= roman_dict[roman[i+1]]:
            result += roman_dict[roman[i]]
        # Si el símbolo actual es menor que el siguiente, lo restamos al resultado
        else:
            result -= roman_dict[roman[i]]

    # Devolvemos el resultado
    return result

# Pruebas de la función
print(decimal_to_roman(119))
print(roman_to_decimal("XIX"))
```

# Oposiciones

```python
#Crea una gráfica que muestre la probabilidad de aprobar unas oposiciones (p.ej., de secundaria), preparando un subconjunto x de N temas totales, considerando que hay b bolas sorteadas.
import matplotlib.pyplot as plt
from math import factorial as f
import scipy
def opos(n,b): # n = numero temas totales, b = numero bolas extraidas
    x = range(n-b)
    y = []
    # X: numero de temas preparados
    for X in x:
        y.append(100*(1. - ( f(n-X)*f(n-b)/(f(n)*f(n-X-b)) )))
    return x,y

def np_opos(n,b):
    x = np.arange(n-b)
    def f(x):
        return scipy.special.factorial(x)
    y = 100*(1. - ( f(n-x)*f(n-b)/(f(n)*f(n-x-b)) ))
    return x,y

def dibuja(n,b):
    fig, ax1 = plt.subplots()
    ax2 = fig.add_axes([0.43, 0.2, 0.4, 0.4]) # inset for zoom
    x, y = opos(n,b)
    ax1.plot(x,y,'o-',label='n='+str(n)+' b='+str(b))
    ax1.set_title("Probabilidad para n temas y b bolas")
    ax1.set_xlabel("Temas preparados")
    ax1.set_ylabel("Probabilidad (%)")
    ax1.legend()
    if n>15:
        # zoom
        ax2.set_title("Zona crítica")
        ax2.plot(x[5:15],y[5:15],'o-')
        ax2.grid('on')
    return
dibuja(69,5)
```
